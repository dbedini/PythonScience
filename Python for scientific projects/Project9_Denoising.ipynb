# Running-mean filter

Smoothes the function by taking the average of k points adiacent to the current one



```python
# Creating noisy signal
import numpy as np
import matplotlib.pyplot as plt
import copy

Nt = 10001
t = np.linspace(0, 4*np.pi, Nt)
noise = np.random.normal(size=Nt)
signal = np.zeros(Nt)
signal_noisy = np.zeros(Nt)
for i in range(1,4):
    signal += np.cos(i*t)**i
signal_noisy = signal + noise

plt.xlabel('time (rad)')
plt.ylabel('amplitude')
plt.title('Noisy signal')
plt.plot(t, signal_noisy, label='noisy')
plt.plot(t, signal, label='original')
plt.legend()
plt.show()
```

```python
# Applying running-mean filter
k = 10
signal_filt = np.zeros(Nt)
for i in range(k,Nt-k):
    signal_filt[i] = np.mean(signal_noisy[i-k:i+k])

plt.plot(t,signal_noisy,label='Noisy')
plt.plot(t,signal_filt,label='Filtered')
plt.xlabel('time (rad)')
plt.ylabel('amplitude')
plt.title('effect of running-mean filtering')
plt.legend()
plt.show()
```

Increasing the number of adiacent points in the average the filtered signal gets closer to the noisless one

```python
# Analizing correlation of filtered and noisy signal 
signal_noisy = signal + np.random.randn(len(signal))

def signal_filt(signal, k):
    N = len(signal)
    sf = copy.deepcopy(signal)
    for i in range(N):
        # mean extremes need to be taken carefully 
        lowerBound = np.max((0,i-k))
        upperBound = np.min((N,i+k))
        sf[i] = np.mean(signal[lowerBound:upperBound])
    return sf

kvals = np.arange(2,50)
corr_coeffs = np.zeros(len(kvals)) 
for k,i in zip(kvals, range(len(kvals))):
    sf = signal_filt(signal_noisy, k)
    corr_coeffs[i] = np.corrcoef(sf, signal)[0,1]
    
plt.plot(kvals, corr_coeffs, 'ks-')
plt.title('Correlation vs number adjacent points')
plt.xlabel('k')
plt.ylabel('corr coeff')
plt.show()
```

Previous consideration confirmed by analysis of correlation coeffs

# Gaussian convolution

A weighted average given by the kernel function that in this case is a gaussian

```python
from scipy.signal import detrend 
import numpy as np
import matplotlib.pyplot as plt
import copy

fsampling = 512
t = np.arange(-2,2,1/fsampling)
Np = len(t)
noise = np.random.normal(scale=1.1, size=Np)
signal = detrend(t**3 + np.sign(t))
signal_noisy = signal + noise

plt.plot(t, signal_noisy, t, signal)
plt.legend(['noisy', 'original'])
plt.show()
```

```python
from scipy.ndimage import gaussian_filter

signal_filt = gaussian_filter(signal_noisy, sigma=5)
plt.plot(t, signal_noisy, t, signal_filt, t, signal)
plt.legend(['noisy','filtered','original'])
plt.show()
```

Parameters of gaussian convolution : number of points, std. variation.

Evaluation of optimal parameters via residual sum of squares



```python
from scipy.signal import detrend 
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import copy

Nf = 100
Nstd = 100 
fs = 512
freqs = np.linspace(0.1, fs, Nf)
stds = np.linspace(0, 1, Nstd)
rss = np.zeros((Nstd,Nf)) # residual sum of squares

for f,i in zip(freqs,range(Nf)):
    for std,j in zip(stds,range(Nstd)):
        t = np.arange(-2,2,1/f)
        noise = np.random.normal(scale=1.1, size=len(t))
        signal = detrend(t**3 + np.sign(t))
        signal_noisy = signal + noise
        signal_filt = gaussian_filter(signal_noisy, sigma=std)
        rss[j,i] = sum((signal_filt-signal)**2)
```

```python
plt.imshow(rss)
Np = 10
plt.xticks(ticks=np.linspace(0, Nstd, Np), labels=np.round(np.linspace(0, stds[-1], Np),1))
plt.yticks(ticks=np.linspace(0, Nf, Np), labels=np.linspace(0.1, freqs[-1], Np).astype('int32'))
plt.ylabel('sampling freq (Hz)')
plt.xlabel('std. dev')
plt.title('Optimization by RSS analysis')
plt.colorbar(label='residual sum of squares')
plt.show()
```

As expected, large std. dev lead to small error only if sampling freq is large
