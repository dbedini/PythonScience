```python
x = np.linspace(-3,3,100)
plt.plot(x,x)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlim(min(x), max(x))
plt.ylim([-10,10])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('plot')
plt.gca().set_aspect(1/plt.gca().get_data_ratio()) # plot diventa quadrato usando ratio x/y
plt.legend(['x','x**2','x**3'])
plt.show()
```

```python
x = np.linspace(-3,3,100)
fig, ax = plt.subplots()
ax.plot(x,x)
ax.plot(x,x**2)
ax.plot(x,x**3)
ax.set_xlim(min(x), max(x))
ax.set_ylim([-10,10])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('plot')
ax.set_aspect(1/ax.get_data_ratio())
ax.grid()
plt.show()
```

```python
x = np.linspace(-10,10,50)
y = x**2
fig,ax = plt.subplots()
ax.plot(x,y, 'ko-', markerfacecolor='w', markersize=5)
ax.annotate('pippo', (x[2],y[2]), fontsize=10, arrowprops=dict(), xycoords='data',xytext=(0,10), horizontalalignment='center')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

n = 200
D = np.zeros((n,2))

D[:,0] = np.linspace(0,10,n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n)*10
sns.jointplot(D[:,0], D[:,1])
plt.show()
```

```python
df = pd.DataFrame(data=D, columns=['var1', 'var2'])
sns.jointplot(x=df.columns[0], y=df.columns[1], data=df, kind='kde')
plt.show()
```

```python
x = np.linspace(-1,1,n)

y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

sns.scatterplot(x=y1, y=y2, hue=y3, legend=False, palette='jet')
plt.show()
```
